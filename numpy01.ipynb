```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
salario1 = np.array([100,200,300,550,250,120])
salario2 = np.array([600,350,400,250,150,100])
plt.plot(salario1, ls='--', marker='s', ms=8, label='Salário Henrique')
plt.plot(salario2, ls='--', marker='^', ms=8, label='Salário Fulano')
plt.legend()
plt.show()
```
